# Время жизни нейтрона

Нейтрон  &mdash; тяжёлая элементарная частица, не имеющая электрического заряда. Нейтроны и протоны являются двумя главными компонентами атомных ядер; общее название для протонов и нейтронов &mdash; нуклоны. 



<center>Рис. 1. Кварковая структура нейтрона</center>

Считается надёжно установленным, что нейтрон является связанным состоянием трёх кварков: одного "верхнего" (*u*) и двух "нижних" (*d*) кварков (кварковая структура *udd*). Близость значений масс протона и нейтрона обусловлена свойством приближённой изотопической инвариантности: в протоне (кварковая структура *uud*) один *d*-кварк заменяется на *u*-кварк, но поскольку массы этих кварков очень близки, такая замена слабо сказывается на массе составной частицы.
Про кварковую модель и трудности измерения массы одного из кварков можно прочитать в блокноте [Масса *t*-кварка](TopQuark.ipynb).



Поскольку нейтрон тяжелее протона (на 0.001 388 449 33(49) а.е.м.), то он может распадаться в свободном состоянии.
Бета-распад нейтрона &mdash; спонтанное превращение свободного нейтрона в протон с излучением $\beta$-частицы (электрона) и электронного антинейтрино:

$$
n \rightarrow p + e^{-}+ \overline{\nu}_{e}.
$$

Диаграмма Фейнмана для бета-распада нейтрона на протон, электрон и электронное антинейтрино при участии виртуального тяжёлого $W^{-}$-бозона приведена на рисунке 2.



<center>Рис. 2. Бета-распад нейтрона</center>

На рисунке 2 вертикальной стрелкой показано направление хода времени. У основания стрелки представлено обозначение начальной частицы &mdash; нейтрона и его кварковой структуры. В верхней части диаграммы &mdash; конечные продукты реакции. Горизонтальной волнистой линией обозначена виртуальная частица &mdash; $W^{-}$-бозон, которая не регистрируется экспериментально.

```octave
# Загрузка пакета для интервальных вычислений
pkg load interval
```

```octave
# Загрузить данные 
load data/neutron.mat

data = neutron.data;
data_vars = neutron.vars

neutron_mid = data(:,1); 
neutron_rad_std = data(:,2); 
neutron_rad_std_syst = data(:,2) + data(:,4); 

# Интервальные измерения только со случайной погрешностью
NTstd = midrad(neutron_mid, neutron_rad_std)

# Интервальные измерения со случайной и систематической погрешностью
NTtotal = midrad(neutron_mid, neutron_rad_std_syst)
```

```octave
# Графическое представление интервальных данных
nums = 1:numel(NTtotal);
vals = mid(NTtotal);
errorbar (vals, nums, rad(NTtotal), rad(NTtotal),">.r");
ylim([0 nums(end)+1])
hold on
h = errorbar (vals, nums, rad(NTstd), rad(NTstd),">.b");
set(h,"linewidth",6)
set(h,"markersize",12)
title ("Interval measurements of neutron life time");
xlabel("Neutron life time, s")
ylabel("Measurement number")
legend ("Measurements with total errors","Measurements with std errors","location","southeast");
#grid on
```

```octave
# Пересечение интервалов пусто, т.е. интервальная выборка несовместна
intersect(NTtotal)
```

```octave
# Функция построения оценки константы по интервальным измерениям 
# методом центра неопределённости (МЦН) [13]
function [oskorbin_center, k] = estimate_uncertainty_center(intervals)
  n = numel(intervals);
  y = mid(intervals);
  e = rad(intervals);
  A = [ones(n,1), -e; -ones(n,1), -e];
  b = [y; -y];  
  c = [0 1];
  lb = [-Inf 0];
  ub = [Inf Inf];  
  ctype(1:2*n) = 'U'; 
  vartype(1:2) = 'C';
  sense = 1;
  [beta, fval, exitcode] = glpk (c, A, b, lb, ub, ctype, vartype, sense);
  if exitcode < 0
      return
  end   
  oskorbin_center = beta(1);
  k = beta(2);
end 
```

```octave
# МЦН-оценка массы t-кварка и минимальный коэффициент расширения интервалов для совместности выборки
[oskorbin_center, k] = estimate_uncertainty_center(TQtotal)
```

```octave
# Графическое представление интервальных данных и МЦН-оценки значения массы t-кварка
errorbar (vals, nums, k*rad(NTtotal), k*rad(NTtotal),">.k");
ylim([0 nums(end)+1])
hold on
h = errorbar (vals, nums, rad(NTtotal), rad(NTtotal),">.r");
set(h,"linewidth",4)
h = errorbar (vals, nums, rad(NTstd), rad(NTstd),">.b");
set(h,"linewidth",6)
set(h,"markersize",12)
title ("Interval measurements of t-quark mass");
xlabel("Neutron life time, s")
ylabel("Measurement number")
plot(oskorbin_center*[1 1], ylim, 'm--')
legend ("k-scaled measurements", "Measurements with total errors",...
        "Measurements with standard errors", "Oskorbin center",...
        "location","southeast");
```
